```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
file_path = r'C:\Users\raipr\Downloads\nse_all_stock_data (1).csv\nse_all_stock_data (1).csv' 
df = pd.read_csv(file_path)
```

```python
missing_values = df.isnull().sum()
```

```python
print("Columns with missing values:")
print(missing_values[missing_values > 0])
```

```python
# Iterate over each column
for column in df.columns:
    # Check if the column has missing values
    if df[column].isnull().any():
        # Calculate the mean of the column
        mean_value = df[column].mean()
        # Fill missing values with the mean
        df[column].fillna(mean_value, inplace=True)
```

```python
descriptive_stats = df.describe()
print(descriptive_stats)
```

```python
# List of columns to drop based on stocks
columns_to_drop = ['COOLCAPS',  'ALLETEC']

# Drop specified columns from the DataFrame
df.drop(columns=columns_to_drop, inplace=True)

# Verify remaining columns in the DataFrame
print(df.columns)
```

```python
# Assuming df is your DataFrame with columns as described
# Example: Statistical analysis of stock prices

# Drop 'Date' column if it's not needed for analysis
df.drop(columns=['Date'], inplace=True)

# Statistical analysis of stock prices
price_stats = df.describe()

# Display the statistics
print("Statistical Analysis of Stock Prices and Volumes:")
print(price_stats)
```

```python
import matplotlib.pyplot as plt

# Select the first 10 stocks for visualization
top_10_stocks = df.columns[:10]

plt.figure(figsize=(12, 8))

# Plotting price trends for each stock
for stock in top_10_stocks:
    plt.plot(df.index, df[stock], label=stock)

plt.title('Price Trends of the First 10 Stocks')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate daily returns
df_returns = df.pct_change().dropna()

# Create correlation matrix of returns
correlation_matrix = df_returns.corr()

# Display correlation matrix
print("Correlation Matrix of Stock Returns:")
print(correlation_matrix)
```

```python
import seaborn as sns

# Find top correlated pairs
stacked_corr = correlation_matrix.stack().sort_values(ascending=False)
top_pairs = stacked_corr[stacked_corr != 1].head(10)  # Adjust number as needed

print("\nTop Correlated Pairs of Stocks:")
print(top_pairs)

# Plot heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Stock Returns')
plt.show()
```

```python
import pandas as pd
from scipy.stats import zscore


df_zscore = df.copy() 

for column in df.columns[1:]:  
    df_zscore[column + '_zscore'] = zscore(df[column])


threshold = 3
anomalies = []

for column in df.columns[1:]:  
    stock_anomalies = df_zscore[(df_zscore[column + '_zscore'] > threshold) | (df_zscore[column + '_zscore'] < -threshold)]
    
    
    anomalies.append(stock_anomalies)


anomalies_df = pd.concat(anomalies)


top_anomalies = anomalies_df.head(5)
print("Top 5 Significant Anomalies:")
print(top_anomalies)


```
